```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from pysolar.solar import get_altitude
from sklearn.metrics import mean_squared_error
from scipy.optimize import curve_fit  # Import curve_fit for optimization

# Define the file paths and their corresponding custom labels
file_paths = {
    'data/long/22-10-2024-10-54-12-cloudy-long.csv': 'Cloudy BLKT Long',
    'data/long/17-10-2024-10-12-46-cloudy-hyde-long.csv': 'Cloudy Hyde Park Long',
    'data/long/15-10-2024-11-07-54-cloudy-as.csv': 'Cloudy AS Long',
}

# Read each file into a DataFrame and store it in a dictionary with the custom label
dataframes = {}
for file_path, label in file_paths.items():
    df = pd.read_csv(file_path, delimiter=r'\s+')
    dataframes[label] = df

# Load the data for the "Cloudy Hyde Park Long" session
selected_label = 'Cloudy Hyde Park Long'
selected_df = dataframes[selected_label]

# Ensure the column names are consistent
selected_df.columns = selected_df.columns.str.strip()

# Calculate solar zenith angle for each time point
latitude = 51.5054772
longitude = -0.1766496
y_data = []

for time_str in selected_df['Time']:
    # Convert time string to datetime object
    time_obj = datetime.datetime.strptime(time_str, '%H:%M:%S').replace(
        year=2024, month=10, day=11, tzinfo=datetime.timezone.utc
    ) - datetime.timedelta(hours=1)
    altitude = get_altitude(latitude, longitude, time_obj)
    zenith_angle = 90 - altitude
    y_data.append(zenith_angle)

# Add the solar zenith angle to the DataFrame
selected_df['SolarZenithAngle'] = y_data

# Resample the data to 1-minute averages
selected_df['Time'] = pd.to_datetime(selected_df['Time'], format='%H:%M:%S')
selected_df.set_index('Time', inplace=True)
selected_df = selected_df.resample('1T').mean().dropna()

# Calculate direct flux for the cloudy condition
total_flux = selected_df['KZ11W:']
diffusive_flux = 0.01 * selected_df['KZ11W:']
direct_flux_cloudy = total_flux - diffusive_flux

# Calculate and print the average direct flux
average_direct_flux_cloudy = direct_flux_cloudy.mean()
print('Average direct flux for cloudy conditions is', average_direct_flux_cloudy, 'W/m^2')

# Define the Haurwitz model for clear sky direct flux
def haurwitz_model(sza):
    """
    Calculate clear sky direct flux using the Haurwitz model.
    sza: Solar Zenith Angle in degrees
    """
    # Convert SZA to radians
    sza_rad = np.radians(sza)
    # Ensure that the cosine of the zenith angle is not negative
    cos_sza = np.maximum(np.cos(sza_rad), 0)
    # Haurwitz model calculation
    flux_clear = 1098 * cos_sza * np.exp(-0.057 / cos_sza)
    return flux_clear

# Calculate the cloud transmittance and optical depth for each 1-minute interval
transmittances = []
optical_depths = []

# Loop through each row in the selected DataFrame
for index, row in selected_df.iterrows():
    # Assuming the solar zenith angle is given in the DataFrame
    sza = row['SolarZenithAngle']

    albedo = 0.39

    # Calculate direct flux in clear sky using Haurwitz model
    direct_flux_clear = haurwitz_model(sza)

    # Calculate cloud transmittance
    transmittance = direct_flux_cloudy.loc[index] / direct_flux_clear
    transmittances.append(transmittance)

    # Calculate cloud optical depth
    a_tau = 0.58
    b_tau = 0.74
    c = 0.1365
    d = 0.1291
    cos_sza = np.cos(np.radians(sza))
    numerator = a_tau + b_tau * cos_sza - transmittance
    denominator = c - d * albedo
    optical_depth = numerator / denominator
    optical_depths.append(optical_depth)

# Add results to the DataFrame
selected_df['CloudTransmittance'] = transmittances
selected_df['CloudOpticalDepth'] = optical_depths

# Calculate average values
average_transmittance = selected_df['CloudTransmittance'].mean()
average_optical_depth = selected_df['CloudOpticalDepth'].mean()

# Print results
print(f"Average cloud transmittance: {average_transmittance:.3f}")
print(f"Average cloud optical depth: {average_optical_depth:.2f}")

# Optional: Save the updated DataFrame to a new CSV
selected_df.to_csv('updated_cloudy_hyde_long.csv', index=False)

# Define the empirical model function with I_0 fixed at 1376
def empirical_flux_model_fixed_I0(X, k, a):
    """
    Empirical model with I_0 fixed at 1376.
    """
    sza, cod = X
    I_0 = 1376  # Fixed value
    # Convert SZA to radians
    sza_rad = np.radians(sza)
    # Ensure that the cosine of the zenith angle is not negative
    cos_sza = np.maximum(np.cos(sza_rad), 0.01)  # Avoid division by zero
    # Empirical model calculation
    flux_empirical = I_0 * cos_sza * np.exp(-k / cos_sza) * np.exp(-a * cod)
    return flux_empirical

# Prepare the data for curve fitting
# Remove any NaN or infinite values to prevent errors during fitting
valid_indices = (
    ~np.isnan(selected_df['SolarZenithAngle']) &
    ~np.isnan(selected_df['CloudOpticalDepth']) &
    ~np.isnan(direct_flux_cloudy)
)
sza_data = selected_df['SolarZenithAngle'][valid_indices]
cod_data = selected_df['CloudOpticalDepth'][valid_indices]
flux_data = direct_flux_cloudy[valid_indices]

# Stack the independent variables
X_data = np.vstack((sza_data, cod_data))

# Initial guesses for k and a
initial_guess = [0.057, 0.1]

# Perform curve fitting to optimize k and a
popt, pcov = curve_fit(
    empirical_flux_model_fixed_I0, X_data, flux_data, p0=initial_guess
)

# Extract the optimized parameters
k_opt, a_opt = popt
print(f"Optimized parameters with I_0 fixed at 1376:\nk = {k_opt:.4f}, a = {a_opt:.4f}")

# Calculate the modeled flux using the optimized parameters
empirical_fluxes_optimized = empirical_flux_model_fixed_I0(
    (selected_df['SolarZenithAngle'], selected_df['CloudOpticalDepth']), k_opt, a_opt
)

# Add the optimized empirical flux to the DataFrame
selected_df['EmpiricalFluxOptimized'] = empirical_fluxes_optimized

# Function to check the accuracy of the optimized empirical model
def check_model_accuracy(actual_flux, predicted_flux):
    """
    Calculate the accuracy of the empirical model by comparing actual flux to predicted flux.
    """
    mse = mean_squared_error(actual_flux, predicted_flux)
    rmse = np.sqrt(mse)
    return rmse

# Calculate the RMSE between the actual direct flux and the optimized empirical flux
rmse_optimized = check_model_accuracy(
    direct_flux_cloudy[valid_indices], selected_df['EmpiricalFluxOptimized'][valid_indices]
)
print(f"RMSE between actual and optimized empirical flux: {rmse_optimized:.2f} W/m^2")

# Plotting the results
plt.figure(figsize=(14, 6))

# Calculate the uncertainty for each measured flux
uncertainties = []

for flux in direct_flux_cloudy:
    zero_offset = 2.5  # W/m^2
    non_linearity_error = 0.03 * flux  # 3% of measured value
    signal_processing_error = 3  # W/m^2
    directional_response_error = 20  # W/m^2

    # Total uncertainty using RSS method
    total_uncertainty = np.sqrt(
        zero_offset**2 +
        non_linearity_error**2 +
        signal_processing_error**2 +
        directional_response_error**2
    )
    uncertainties.append(total_uncertainty)

# Convert uncertainties to a NumPy array
uncertainties = np.array(uncertainties)

# Plot Time Averaged Direct Flux with error bars
plt.errorbar(
    selected_df.index, direct_flux_cloudy, yerr=uncertainties, fmt='o',
    label='Measured Direct Flux (Cloudy)', color='r', ecolor='gray', capsize=3
)

# Plot Optimized Empirical Flux
plt.plot(
    selected_df.index, selected_df['EmpiricalFluxOptimized'],
    label='Optimized Empirical Flux', color='orange', linestyle='--'
)

# Plot Time Averaged Direct Flux
# plt.plot(selected_df.index, direct_flux_cloudy, label='Time Averaged Direct Flux (Cloudy)', color='r')
# Set plot labels and title
plt.xlabel('Time')
plt.ylabel('Flux (W/m)')
plt.title('Flux vs. Time with Optimized Empirical Model (I = 1376) and Instrument Uncertainties')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
