```python
cloud_base_height = 1000 # use online data 
cloud_top_height = 2000 # use online data
thickness = cloud_top_height - cloud_base_height

def calculate_liquid_water_path(thickness, optical_depth): # use mie scattering theory
    density_water = 1000 # kg/m^3
    mean_drop_radius = 10e-6 # m
    liquid_water_path = 2 * density_water * mean_drop_radius * optical_depth /3 
    return liquid_water_path

def calculate_liquid_water_content(liquid_water_path, thickness): # mixing ratio of liquid water to air
    liquid_water_content = liquid_water_path / thickness
    return liquid_water_content

def calculate_surface_water_vapour_mixing_ratio(T_C, relative_humidity, atmospheric_pressure):
    """
    kg/kg

    
    T_C (float): 
    RH (float): 0-100
    P (float):  hPa

    
    float: kg/kg
    """
    # 
    a = 23.5518
    b = 2937.4
    c = 4.9283
    epsilon = 0.622 # 

    # 
    T_K = T_C + 273.15

    #  (Psat)
    pressure_saturation = 10 ** (a - (b / T_K)) * T_K - c

    #  (e)
    e = (relative_humidity / 100) * pressure_saturation

    #  (w)
    w = epsilon * (e / (atmospheric_pressure - e))

    print(f"pressure_saturation: {pressure_saturation:.2f} hPa")
    print(f"e: {e:.2f} hPa")
    print(f"water mixing ratio: {w:.9f} kg/kg")

    return w

# 
T_C = 20  # 
RH = 70   # %
atmospheric_pressure = 1013.25  # hPa

liquid_water_path = calculate_liquid_water_path(thickness, optical_depth)
liquid_water_content = calculate_liquid_water_content(liquid_water_path, thickness)
surface_water_vapour_mixing_ratio = calculate_surface_water_vapour_mixing_ratio(T_C, RH, atmospheric_pressure)

print(f"Cloud liquid water path: {liquid_water_path:.9f} kg/m^2")
print(f"Cloud liquid water content: {liquid_water_content:.9f} kg/m^3")
print(f"Surface water vapour mixing ratio: {surface_water_vapour_mixing_ratio}")
```

```python
# determine cloud radiative forcing, it is the difference in radiative flux between cloudy and clear sky conditions
# longwave radiation, cloud reduce radiation by absorbing and re-emitting IR 

import numpy as np

def calculate_cloud_radiative_forcing(surface_temperature_celsius, cloud_base_height, cloud_top_height, surface_emissivity=0.98):
    """
    W/m^2

    
    surface_temperature_celsius (float): 
    cloud_base_height (float): 
    cloud_top_height (float): 
    surface_emissivity (float):  0.98

    
    float: W/m^2
    """
    # 
    stefan_boltzmann_constant = 5.67e-8  # Stefan-Boltzmann  (W m^-2 K^-4)
    moist_adiabatic_lapse_rate = 0.006  #  (K/m)

    # 
    surface_temperature_kelvin = surface_temperature_celsius + 273.15

    # 
    surface_longwave_outgoing_flux = surface_emissivity * stefan_boltzmann_constant * surface_temperature_kelvin**4

    # 
    cloud_top_temperature_kelvin = surface_temperature_kelvin - moist_adiabatic_lapse_rate * (cloud_top_height - cloud_base_height)

    # 
    cloud_top_longwave_outgoing_flux = surface_emissivity * stefan_boltzmann_constant * cloud_top_temperature_kelvin**4

    # 
    cloud_radiative_forcing_longwave = surface_longwave_outgoing_flux - cloud_top_longwave_outgoing_flux

    return cloud_radiative_forcing_longwave

# 
surface_temperature_celsius = 20  #  ()
cloud_base_height = 500    #  (m)
cloud_top_height = 1500    #  (m)

cloud_radiative_forcing = calculate_cloud_radiative_forcing(surface_temperature_celsius, cloud_base_height, cloud_top_height)
print(f" (CRF_LW): {cloud_radiative_forcing:.2f} W/m^2")



```
