```python
# Plot to calculate vertical column optical depth of clear sky
# Vertical column optical depth - Beer-Lambert-Bouguer Law
# I = I0 * exp(-u*sec(zenith_angle))
# ln(I) = -u*sec(zenith_angle) + ln(I0)

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pysolar.solar import get_altitude
import datetime

# Load the data for the "Both Up Session"
selected_label = 'Both Up Session'
selected_df = dataframes[selected_label]

# Calculate direct flux
total_flux = selected_df['KZ11W:']
diffusive_flux = 0.01 * selected_df['KZ11W:']
direct_flux = total_flux - diffusive_flux

# Calculate solar zenith angle for each time point
latitude = 51.5054772
longitude = -0.1766496
y_data = []

for time_str in selected_df['Time']:
    # Convert time string to datetime object
    time_obj = datetime.datetime.strptime(time_str, '%H:%M:%S').replace(year=2024, month=10, day=11, tzinfo=datetime.timezone.utc)
    altitude = get_altitude(latitude, longitude, time_obj)
    zenith_angle = 90 - altitude
    y_data.append(zenith_angle)

# Discard the last data point
sec_zenith_angle = 1 / np.cos(np.radians(y_data[:-1]))
ln_I = np.log(direct_flux[:-1])

# Plotting
plt.plot(sec_zenith_angle, ln_I, 'o-')
plt.xlabel('sec(zenith_angle)')
plt.ylabel('ln(I)')
plt.title('ln(I) vs sec(zenith_angle)')

# Calculate the gradient (vertical column optical depth)
gradient, intercept = np.polyfit(sec_zenith_angle, ln_I, 1) 

# Plotting the fitted line
plt.plot(sec_zenith_angle, gradient * np.array(sec_zenith_angle) + intercept, 'r--')

# Display the gradient value on the plot
optical_depth = -gradient
plt.text(sec_zenith_angle[0], ln_I[0], f'Gradient: {optical_depth:.2f}', color='red')

plt.grid(True)
plt.show()

for time_str, angle, intensity in zip(selected_df['Time'][:-1], y_data[:-1], direct_flux[:-1]):
    print(f"Time: {time_str}, Zenith Angle: {angle:.2f}, Intensity: {intensity:.2f}")
```

```python
# determine cloud radiative forcing, it is the difference in radiative flux between cloudy and clear sky conditions
# longwave radiation, cloud reduce radiation by absorbing and re-emitting IR 

def calculate_cloud_radiative_forcing(total_flux_clear, total_flux_cloudy):
    cloud_radiative_forcing = total_flux_clear - total_flux_cloudy
    return cloud_radiative_forcing

```
